## 🔎 What Are Higher-Order Functions?
A **higher-order function** is a function that:  
1. Takes another function as an argument, OR  
2. Returns a function as its result.  

💡 **in Simple Terms**  
It’s a function that works *with other functions*.  

## 🍴 Real-Life Examples
- **Organization:** A manager doesn’t do tasks themselves. They assign tasks to employees.  
- **Restaurant:** A head chef doesn’t cook every dish. They delegate tasks to junior chefs.  

➡️ Just like that, higher-order functions **delegate work to other functions**.

## 🛠️ Common Higher-Order Functions in Python

### 1. `map()`
Applies a function to each element in an iterable.

````python
def double(x):
    return x * 2

numbers = [1, 2, 3, 4, 5]
doubled = list(map(double, numbers))
print(doubled)
````

### 2. `filter()`
Keeps only the elements that match a condition.

````python
def is_even(x):
    return x % 2 == 0

numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(is_even, numbers))
print(evens)
````

### 3. `reduce()`
Applies a function cumulatively to items in an iterable.  
⚠️ Must be imported from `functools`.

````python
from functools import reduce

def multiply(x, y):
    return x * y

numbers = [1, 2, 3, 4, 5]
result = reduce(multiply, numbers)
print(result)
````

### 4. `zip()`
Combines multiple iterables element by element.

````python
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]

zipped = list(zip(names, ages))
print(zipped)
````

### 5. `enumerate()`
Adds an index to each item in an iterable.

````python
fruits = ['apple', 'banana', 'cherry']

for index, fruit in enumerate(fruits, start=1):
    print(f"{index}. {fruit}")
````

## ⚡ Lambda Functions
A **lambda function** is a **small, anonymous function**.  
It can take any number of arguments, but it only has **one expression**.

📌 Syntax:
```python
lambda arguments: expression
```

````python
numbers = [1, 2, 3, 4, 5]

squared = list(map(lambda x: x**2, numbers))
print(squared)
````

## 🛠️ Mini-Project: Data Transformation Pipeline

**Task:** Clean messy student data using higher-order functions.

````python
from functools import reduce

students = ["  alice ", "BOB", "ChArLiE  ", "dave"]

# Step 1: Clean and format names
cleaned = list(map(lambda x: x.strip().title(), students))

# Step 2: Keep names longer than 3 characters
filtered = list(filter(lambda x: len(x) > 3, cleaned))

# Step 3: Count total characters
total_chars = reduce(lambda x, y: x + len(y), filtered, 0)

# Step 4: Assign IDs
with_ids = list(enumerate(filtered, start=1))

print("Cleaned Names:", cleaned)
print("Filtered Names:", filtered)
print("Total Characters:", total_chars)
print("Student IDs:", with_ids)
````

## ⚠️ Common Mistakes & Debugging Checklist
- ❌ Forgetting `from functools import reduce`  
- ❌ Expecting `map()` or `filter()` to give lists → need `list()`  
- ❌ Writing overly complicated lambdas (hard to read)  
- ❌ Mixing up function reference vs call (`func` vs `func()`)
