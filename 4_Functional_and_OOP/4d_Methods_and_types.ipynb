## 🔍 What Are Methods?

In Python, **methods** are functions defined inside a class. These allow objects (instances) to *do something* or *behave in specific ways*.

> Think of a class as a blueprint for making real-world things.  
> Methods are like the *instructions* or *actions* that each object can perform.

There are **3 main types of methods**:

| Type           | First Parameter | Access Instance Data | Access Class Data | Typical Use                        |
|----------------|------------------|------------------------|--------------------|-------------------------------------|
| Instance Method| `self`           | ✅ Yes                | ❌ No             | Modify or use instance attributes  |
| Class Method   | `cls`            | ❌ No                 | ✅ Yes            | Work with the class itself         |
| Static Method  | (none)           | ❌ No                 | ❌ No             | Utility functions inside class     |


## 🧱 2. Instance Methods

**Instance methods** are the most common type of method in a class.  
They operate on a specific object and always have `self` as the first parameter.

`self` refers to the current instance (object) calling the method.


````python
# 👷 Class that represents a house
class House:
    def __init__(self, color):
        # Store the color as an instance variable
        self.color = color

    def show_color(self):
        # Access and print this object's color
        print(f"This house is painted {self.color}")

# Create an instance of House
my_house = House("blue")

# Call the instance method
my_house.show_color()
````

## 🏗️ 3. Class Methods

**Class methods** work with the class itself, not specific objects.  
They use `cls` (not `self`) as the first parameter.

Use `@classmethod` decorator above the method.

Class methods are great for:
- Making **alternate constructors**
- Working with **class-level data**


````python
class House:
    def __init__(self, color, age):
        self.color = color
        self.age = age

    @classmethod
    def from_construction_year(cls, color, year_built):
        age = 2025 - year_built
        return cls(color, age)

# Use the class method as an alternate constructor
house2 = House.from_construction_year("white", 2000)
print(house2.age)
````

## 🔧 4. Static Methods

**Static methods** don't use `self` or `cls`.  
They're just utility functions grouped inside a class.

Use `@staticmethod` decorator above the method.

They:
- Do **not** access instance or class data
- Are useful for **validation**, **calculations**, or **helpers**


````python
class House:
    def __init__(self, color, area):
        self.color = color
        self.area = area

    @staticmethod
    def is_valid_area(area):
        return area > 0

print(House.is_valid_area(120))   # ✅ True
print(House.is_valid_area(-50))   # ❌ False
````

## ⚖️ Methods vs Functions

| Feature    | Function                     | Method                              |
|------------|------------------------------|--------------------------------------|
| Location   | Outside class                | Inside class                         |
| Call style | `func_name()`                | `object.method()`                    |
| Use `self`?| ❌ No                         | ✅ Instance method uses `self`       |
| Use `cls`? | ❌ No                         | ✅ Class method uses `cls`           |
| Purpose    | General purpose actions      | Behavior tied to objects or classes  |

> A method is just a function *inside* a class!


## 🧮 Combined Example: `MathTool` Class


````python
class MathTool:
    pi = 3.14159

    def __init__(self, radius):
        self.radius = radius

    # Instance method
    def area(self):
        return MathTool.pi * self.radius ** 2

    # Class method
    @classmethod
    def from_diameter(cls, diameter):
        return cls(diameter / 2)

    # Static method
    @staticmethod
    def is_valid_radius(radius):
        return radius > 0

# ✅ Use instance method
tool = MathTool(5)
print("Area:", tool.area())

# ✅ Use class method
tool2 = MathTool.from_diameter(10)
print("Radius from diameter:", tool2.radius)

# ✅ Use static method
print("Is radius valid?", MathTool.is_valid_radius(-1))
````

```python
class StudentScore:
    # Constructor (__init__) runs when an object is created
    def __init__(self, name, score):
        self.name = name              # Save student name to the instance
        self.score = score            # Save student score to the instance

    # Instance method to check if student passed
    def is_passed(self):
        # Check if score is 50 or more
        return self.score >= 50

    # Class method that allows creating an instance using a grade (A, B, C)
    @classmethod
    def from_grade(cls, name, grade):
        # Dictionary to convert grades to numeric scores
        grade_to_score = {
            'A': 90,
            'B': 75,
            'C': 60
        }
        # Convert the grade to uppercase and get its score (default to 0 if not found)
        score = grade_to_score.get(grade.upper(), 0)
        # Return a new object of the class using the converted score
        return cls(name, score)

    # Static method to check if score is between 0 and 100
    @staticmethod
    def is_valid_score(score):
        return 0 <= score <= 100


# Create an object using direct score
s1 = StudentScore("Ali", 82)  # Name: Ali, Score: 82
print(f"{s1.name} passed? {s1.is_passed()}")  # Should print: Ali passed? True

# Create an object using grade instead of score
s2 = StudentScore.from_grade("Sara", "B")     # Grade B → Score 75
print(f"{s2.name}'s score from grade: {s2.score}")  # Should print: Sara's score from grade: 75
print(f"{s2.name} passed? {s2.is_passed()}")        # Should print: Sara passed? True

# Using static method to check if a score is valid
print("Is 105 a valid score?", StudentScore.is_valid_score(105))  # False (out of range)
print("Is 88 a valid score?", StudentScore.is_valid_score(88))    # True (within range)

## ⚠️ Common Mistakes to Avoid

- ❌ Forgetting `self` in instance methods → Causes runtime error
- ❌ Using `self` instead of `cls` in class methods
- ❌ Using static method when you actually need instance data
- ❌ Calling a method without parentheses (e.g., `obj.method` instead of `obj.method()`)
- ❌ Trying to access instance variables inside static method (will fail!)
