## 🧠 What is Object-Oriented Programming?
Object-Oriented Programming (OOP) is a way to **organize code** by bundling data and behavior into one structure called a class.

A **class**  acts like a blueprint that can create many **objects** , and each object has its own data and behavior.

## 🏗️ Classes and Objects
- A **class** is like a **blueprint** or **template**. It defines what data (**attributes**) and actions (**methods**) something should have.  
- An **object** is a **real-world version** created using that blueprint. It contains its own data and can perform the defined actions.


### 🧩 Analogy: Cookie Cutter
Think of a **class** as a cookie cutter and an **object** as a cookie. You use the cutter (class) to make many cookies (objects), each possibly with different decorations.

## 🔧 Defining a Class and Creating an Object

````python
class Robot:
    def __init__(self, name, color):
        self.name = name        # attribute
        self.color = color      # attribute

# Create an object of Robot
r1 = Robot("RoboOne", "Silver")
print(r1.name)   # Output: RoboOne
print(r1.color)  # Output: Silver
````

## 🛠️ The `__init__()` Method (Constructor)
- Called **automatically** when you create a new object.
- It initializes the object’s attributes.
- `self` refers to the object being created.

### 📌 Visual:
`self.name = name` means: “This object’s `name` is equal to the value passed when creating it.”

## 📬 Accessing Attributes

````python
print("Robot Name:", r1.name)
print("Robot Color:", r1.color)
````

## 👯 Creating Multiple Objects
You can create many objects from the same class, each with different values.

````python
# Define a class named Animal
class Animal:
    # Constructor method to initialize species and sound
    def __init__(self, species, sound):
        self.species = species  # Store the species (e.g., Cat, Dog)
        self.sound = sound      # Store the sound (e.g., Meow, Woof)

# Create first object of Animal class
a1 = Animal("Cat", "Meow")

# Create second object of Animal class
a2 = Animal("Dog", "Woof")

# Print species and sound of the first object
print(f"{a1.species} says {a1.sound}")  # Output: Cat says Meow

# Print species and sound of the second object
print(f"{a2.species} says {a2.sound}")  # Output: Dog says Woof
````

## ⚠️ Common Mistakes
- Forgetting `self` in the `__init__()` method
- Not using dot notation to access attributes (e.g., `obj.attr`)
- Misplacing indentation inside the class
