#### 🧠 What Is Scope?
Think of your program as a big house. Each **room** is like a **function** or a block of code. A variable defined in one room stays in that room unless it's part of the house's central system.

- **Global Variables** = House-wide variables (accessible anywhere)
- **Local Variables** = Room-specific variables (only accessible inside that room/function)

---

### 🌎 Global Scope Example


````python


# Global variable defined at top level
message = "Hello from the global world!"

def show_message():
    # Accessing a global variable
    print(message)

show_message()
print(message)  # Accessible here too
````


📝 **Explanation:**
- `message` is global, so both inside `show_message()` and outside it, we can read its value.

---



### 🔒 Local Scope Example

````python

def greet():
    greeting = "Hi, local world!"  # Local variable
    print(greeting)


greet()
# print(greeting)  # NameError: greeting is not defined
````


📝 **Explanation:**
- `greeting` exists only inside `greet()`. Trying to use it outside causes an error.

---


### 🔄 Modifying Global Variables

````python

counter = 0  # Global variable

def increment():
    global counter  # Tell Python we mean the global variable
    counter += 1

increment()
increment()
print("Counter:", counter)

````



📝 **Explanation:**
- Without `global`, Python would treat `counter` as local and error out.

---

### 🌗 Variable Shadowing

````python

value = 10  # Global

def shadow():
    value = 5  # Local variable shadows global
    print("Inside function, value =", value)

shadow()
print("Outside function, value =", value)
````


📝 **Explanation:**
- Inside `shadow()`, `value` refers to the local one (5). Outside, it’s still 10.

### ⚠️ Common Pitfalls
- **NameError**: Accessing local variables outside their function.
- **UnboundLocalError**: Modifying global variables without `global`.
- **Shadowing**: Accidentally hiding global variables with locals.

---



## ✅ Wrap-Up
- **Global scope** = accessible anywhere in the file
- **Local scope** = accessible only inside the block/function
- Use `global` sparingly—prefer passing values and returning results
- Be mindful of variable names to avoid **shadowing**
