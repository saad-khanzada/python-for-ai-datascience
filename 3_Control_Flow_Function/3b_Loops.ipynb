### 🔁 1. For Loops – Repeating with a Plan
📄 Example: Printing a To-Do List

````python
# A list of tasks
tasks = ["Wake up", "Brush teeth", "Eat breakfast", "Go to school"]

# Looping through the tasks
for task in tasks:
    print("✅ Task:", task)
````

🧠 **Explanation:**  
The `for` loop is going through each item in the list one by one and printing it. Like checking off items in a daily planner.

### ⏳ 2. While Loops – Keep Going Until Told to Stop


🚀 Example: Countdown for Rocket Launch

````python
count = 5
while count > 0:
    print("🚀 Launching in", count)
    count -= 1
print("🌟 Liftoff!")
````

#### 🧠 **Explanation:**  
The loop repeats as long as `count > 0`. When it hits 0, the loop stops, and liftoff begins!

### 🔢 3. Using range() – Looping with Numbers


🎯 Example: Printing Numbers from 1 to 5

````python
for number in range(1, 6):
    print("🎯 Number:", number)
````

🧠 **Explanation:**  
`range(1, 6)` means numbers from 1 to 5 — the end value (6) is excluded.

### 🛑 4. Break – Jump Out of the Loop


🕵️ Example: Secret Word Detector

````python
words = ["apple", "banana", "stop", "grape"]

for word in words:
    if word == "stop":
        print("🛑 'stop' found! Ending loop.")
        break
    print("🔍 Checking:", word)
````

🧠 **Explanation:**  
When the word `"stop"` is found, `break` exits the loop immediately.

### 🌀 5. Continue – Skip and Move On


🧼 Example: Skipping Dirty Dishes

````python
dishes = ["clean", "dirty", "clean", "dirty", "clean"]

for dish in dishes:
    if dish == "dirty":
        print("🧽 Skipping dirty dish...")
        continue
    print("🍽️ Serving:", dish)
````

🧠 **Explanation:**  
When it hits a "dirty" dish, `continue` skips it and moves to the next.

---

### 🧩 6. Else with Loops – The Secret Ending


🧪 Example: Search and Succeed

````python
numbers = [3, 7, 10, 15]
for n in numbers:
    if n == 5:
        print("✅ Found 5!")
        break
else:
    print("❌ 5 not found.")
````

🧠 **Explanation:**  
The `else` block only runs if the loop wasn't broken.

### 🔁🌀 7. Nested Loops – Loops within Loops


🍕 Example: Pizza Orders by Topping and Size

````python
sizes = ["Small", "Medium", "Large"]
toppings = ["Cheese", "Pepperoni"]

for size in sizes:
    for topping in toppings:
        print(f"🍕 Order: {size} {topping} Pizza")
````

🧠 **Explanation:**  
Each size is combined with each topping — total combinations: 3 sizes × 2 toppings = 6 pizzas.
