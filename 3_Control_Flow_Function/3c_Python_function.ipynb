## 🔧 What is a Function?

A function is like a **vending machine**:
- You give it input (like money)
- It performs an action (dispenses snacks)
- It may return a result (a chocolate bar!)

Functions make your code **modular**, **reusable**, and **easier to manage**.

---


## 🔎 Using Built-in Functions

Python has many built-in functions you can use right away:


````python
numbers = [5, 3, 9, 1]
print("Total:", sum(numbers))      # Adds all numbers
print("Length:", len(numbers))     # Counts items
print("Sorted:", sorted(numbers))  # Returns a new sorted list
numbers.sort()                      # Sorts the original list
print("In-place Sort:", numbers)
````

---

````python
## ✍️ Defining Your Own Function


def say_hello():
    """This function prints a friendly greeting."""
    print("Hello, Python beginner!")

# Calling the function
say_hello()
````


🧠 **Explanation:** `def` starts the function. You can call it later using its name.

---


````python
## 🎯 Function with Parameters


def greet(name):
    """Greets the user by name."""
    print("Hello,", name)

greet("Ayesha")
````


🧠 **Explanation:** `name` is a parameter. We pass data to the function.

---


````python
## 🎁 Function with Return Value


def add(a, b):
    """Returns the sum of a and b."""
    return a + b

result = add(5, 7)
print("Result:", result)
````

🧠 **Explanation:** `return` gives back a result we can store or use later.

---

````python
## 🙈 Function Without Return (returns None)


def display_total(a, b):
    print("Total:", a + b)

output = display_total(3, 4)
print("Returned:", output)  # This will print: None


````

## 📜 Documenting with Docstrings

````python

def multiply(a, b):
    """Multiplies two numbers and returns the result."""
    return a * b

help(multiply)
````

## ⚠️ Common Mistakes

- ❌ Forgetting to call the function after defining it
- ❌ Using `print()` instead of `return` for reusable values
- ❌ Not using parentheses when calling a function
- ❌ Re-using the same variable names inside and outside functions (confusing scope)
