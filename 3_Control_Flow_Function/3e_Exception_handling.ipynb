### 🧠 What is an Exception?
An **exception** is like a **problem** that stops your program if it's not handled.
Imagine trying to divide by zero — Python won't allow it, and it'll crash unless you handle the error.

---

````python
## 🔐 The try-except Block

try:
    number = int(input("Enter a number: "))
    print("10 divided by your number is:", 10 / number)
except:
    print("Something went wrong!")

````

🧠 **Explanation:**
- `try`: Python tries this code.
- `except`: If there's any error, run this instead.

### ✅ Using else with try-except

````python
try:
    x = int(input("Enter a number: "))
except ValueError:
    print("Not a number!")
else:
    print("You entered:", x)
````


🧠 **Explanation:**
- `else` runs only **if no exception** occurs.

---

````python

## 🔄 finally Block: Always Runs

try:
    print("Trying something risky...")
    result = 10 / 0
except ZeroDivisionError:
    print("You can't divide by zero!")
finally:
    print("This runs no matter what.")
````


🧠 **Use Case:** Closing a file or connection, cleaning up memory, logging.

---

````python
## 📂 Real-World Example: File Reading

try:
    file = open("nonexistent.txt", "r")
    content = file.read()
    print(content)
except FileNotFoundError:
    print("File not found.")
finally:
    print("Closing file (if opened).")
````

```python
def safe_calculator():
    try:
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        op = input("Choose operation (+, -, *, /): ")
        
        if op == '+':
            result = a + b
        elif op == '-':
            result = a - b
        elif op == '*':
            result = a * b
        elif op == '/':
            result = a / b
        else:
            print("Invalid operator!")
            return

        print("Result:", result)

    except ValueError:
        print("Invalid number input.")
    except ZeroDivisionError:
        print("Can't divide by zero!")
    finally:
        print("Calculation complete.")
safe_calculator()
```
</details>

### ❗ Common Mistakes Beginners Make

- ❌ Forgetting `except` block — crash on error
- ❌ Using a generic `except:` without knowing what it catches
- ❌ Thinking `finally` only runs **if** there's an error (it always runs!)
- ❌ Catching the wrong type (e.g., using `ValueError` for divide-by-zero)
- ❌ Not testing edge cases like 0, or no input
