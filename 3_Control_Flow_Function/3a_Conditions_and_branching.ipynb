### 🌟 What Are Conditions and Branching?

- **Conditions** check whether something is **True** or **False**.
- **Branching** allows your program to **choose a path** based on that condition.

---




#### 🚦 A Simple Analogy:  
**Imagine you are driving on a road and reach a split:**

- 🛣️ **If** you want coffee, **turn left**.
- 🛣️ **Else**, **go right** for tea.

> 🧠 **In programming,** the same idea applies:  
> Your code **decides** what to do based on the condition's result (True or False)!

### 🔹 The if-else Statement

#####  Syntax: `if-else` Statement

```python
if condition:
    # Do something
else:
    # Do something else


##### ➡️ Example 2: Deciding Movie Tickets

````python
# Decide ticket price based on age

age = 15

if age < 18:
    print("You get a child discount!")
else:
    print("You pay the regular price.")
````

#### 🧠 Explanation

- If the person is **younger than 18**, they receive a **discount**.
- **Else**, they pay the **full price**.

---

##### 👉 Real-Life Analogy

> 🎟️ **"Children get cheaper tickets, adults pay full price."**

Just like in real life, the program uses age to decide how much someone should pay.

### 🔹 The if-elif-else Statement


##### Syntax:   `if-elif-else` Statement



```python
if condition1:
    # Do something
elif condition2:
    # Do something else
else:
    # Do something different


##### ➡️ Example 3: Grading System

````python
# Give grades based on marks

marks = 75

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
else:
    print("Grade: F")
````

#### 🧠 Explanation

- The program **checks conditions one by one**, from top to bottom.
- It **stops** as soon as it finds a condition that is **True**.
- Only the **first matching block** of code runs.

---

#### 👉 Real-Life Analogy

> 🧾 **"Depending on your marks, you get different grades."**  
If you score above 90, you get an A.  
Else if above 80, you get a B... and so on.

Just like this, the program **chooses only one path** based on the first condition that is true.

### 🔹 Nested Conditions


Sometimes we need a condition inside another condition.

#### ➡️ Example 4: Bank Loan Eligibility

````python
# Check loan eligibility based on age and income

age = 25
income = 40000

if age >= 18:
    if income >= 30000:
        print("Loan Approved!")
    else:
        print("Loan Denied: Insufficient income.")
else:
    print("Loan Denied: Too young to apply.")
````

##### 🧠 Explanation:
- First, we **check the age**.
- Then, we **check income or another condition** inside the first one.
- This is called **"nested conditions"**, where one `if` is **inside** another `if`.

---

##### 👉 Real-life Analogy:
> 🧾 "First, check if a person is an adult.  
> Then check if they can afford a loan."

Just like that, your program **makes decisions inside decisions**.

### 🔹 Shorthand (One-Line) If-Else


Python lets you write short conditions on a single line.



#### ➡️ Syntax:

```python
output = value_if_true if condition else value_if_false


#### ➡️ Example 5: Quick Discount Decision

````python
# Short version of if-else

age = 17
ticket_price = 5 if age < 18 else 10

print(f"Ticket Price: ${ticket_price}")
````


##### 🧠 Explanation:
- If the **condition is `True`**, the first value is selected.
- Else, the second value is used.
- Perfect for writing **concise decisions in one line**.

---

##### 👉 Real-life Analogy:
> "Quickly decide between two options without writing many lines."

Like:
> ✅ "If you're hungry, grab a snack — otherwise, continue working."

#### ⚠️ Common Mistakes to Avoid

---

##### ❌ Forgetting **colons (`:`)** after `if`, `else`, or `elif` statements.

Make sure to always use **colons** at the end of these statements to mark the start of the code block.

---

##### ❌ **Wrong indentation** – Python uses **spaces** (not tabs) to group code into blocks.

Ensure proper indentation (typically 4 spaces per level) to avoid errors.

---

##### ❌ Using **`=`** instead of **`==`** in conditions.

- **`=`** is for assignment.
- **`==`** is for comparison.
