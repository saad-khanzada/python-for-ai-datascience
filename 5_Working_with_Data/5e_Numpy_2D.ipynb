````python

# 🔵 Create a 3x3 array
arr = np.array([
    [10, 20, 30],   # row 0
    [40, 50, 60],   # row 1
    [70, 80, 90]    # row 2
])

# 🟡 Access specific element at row 1, column 2 (zero-based indexing)
print("Element at (1,2):", arr[1, 2])  # ➤ 60

# 🟢 Get the entire first row (index 0)
print("First row:", arr[0])  # ➤ [10 20 30]

# 🔴 Get the second column (index 1) across all rows
print("Second column:", arr[:, 1])  # ➤ [20 50 80]

# 🟣 Slice from row 1 onward, column 2 onward (bottom-right corner)
print("Last column of last two rows:\n", arr[1:, 2:])  # ➤ [[60], [90]]
````

### 🔹 2D Array Operations – Made Easy 🧠

Now that we have 2D arrays, let’s see how to add, multiply, and even **mix shapes** using NumPy.

We’ll start with two simple 2x2 arrays to make it super clear.

### 📌 Concepts Covered:
- Adding two 2D arrays (same shape)
- Multiplying all values by a number
- Adding a column of values to every row (broadcasting)

Let’s take it step by step 👇


We can also add arrays. The process is identical to matrix addition. Matrix addition of <code>X</code> and <code>Y</code> is shown in the following figure:


<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%205/images/NumTwoAdd.png" width="500">


The numpy array is given by X and Y

````python
# Create a numpy array X

X = np.array([[1, 0], [0, 1]])
X
````

````python
# Create a numpy array Y

Y = np.array([[2, 1], [1, 2]])
Y
````

````python
# Add X and Y

Z = X + Y
Z
````

````python

# 🔵 : Create two small 2D arrays
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# 🟡 NOW : Add them (element-wise addition)
print("A + B (adds each element from A and B):")
print(A + B)  # [ [1+5, 2+6], [3+7, 4+8] ] => [[6, 8], [10, 12]]
````


#### 🧮 What just happened?

We added two **arrays with the same shape**.

Think of them like this:



A + B = Result


**[[1, 2] + [5, 6]** = [6, 8]


**[3, 4]] [7, 8]]**= [10, 12]]


Each number is added to the one in the same position.

✅ This is called **element-wise addition**.

###Scalar multiplication

Multiplying a numpy array by a scaler is identical to multiplying a matrix by a scaler. If we multiply the matrix Y by the scaler 2, we simply multiply every element in the matrix by 2, as shown in the figure.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%205/images/NumTwoDb.png" width="500">


We can perform the same operation in numpy as follows

````python
# Create a numpy array Y

Y = np.array([[2, 1], [1, 2]])
Y
````

````python
# Multiply Y with 2

Z = 2 * Y
Z
````

````python

# Another example 🔴 : Multiply every element in A by 10 (scalar multiplication)
print("A * 10 (each value is multiplied by 10):")
print(A * 10)
````


#### 🔍 This is called scalar multiplication

We’re multiplying the **whole array** by a number:



**A * 10**= Result


**[[1, 2] * 10** = [10, 20]


**[3, 4]] * 10**= [30, 40]]


✅ Every number is scaled (multiplied) by 10.

### 🔹 Axis-Based Math Operations

Use the `axis` parameter:
- `axis=0` → column-wise (down)
- `axis=1` → row-wise (across)




````python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Sum by column:", np.sum(arr, axis=0))
print("Sum by row:", np.sum(arr, axis=1))

````

````python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print("Max of each column:", np.max(arr, axis=0))
print("Mean of each row:", np.mean(arr, axis=1))
````

### 🔹 Useful 2D Functions

These help you understand or reshape your data


````python
arr = np.arange(12).reshape(3, 4)
print("Original Array:\n", arr)

print("Shape:", arr.shape)  # (rows, cols)
````
