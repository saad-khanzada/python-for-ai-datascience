## 🔢 What is Indexing in NumPy?

**Indexing** means selecting a single element from an array using its position (index).

In NumPy (like Python), indexing starts from **0**, not 1.

For example:


````python
# Creating a 1D array
arr = np.array([10, 20, 30, 40, 50])

# Indexing: accessing single elements
print("Element at index 0:", arr[0])   # Output: 10
print("Element at index 3:", arr[3])   # Output: 40

````


You can also use negative indexing to access elements from the end:

For example:
```python
print(arr[-1])  # Output: 40 (last element)


<br>


## ✂️ What is Slicing in NumPy?

**Slicing** means selecting a group (range) of elements from a NumPy array.

The syntax is:  
`array[start : stop]`  
This includes `start` but **excludes** `stop`.

Example:

````python
arr = np.array([10, 20, 30, 40, 50])
print(arr[1:4])  # Output: [20 30 40] → index 1 to 3
````

# 🔹 Array Operations
With NumPy, you can do math on **whole arrays at once** — no loops needed!

---

````python
arr = np.array([1, 2, 3])

# Element-wise multiplication
print(arr * 2)   # 🔵 Doubles each element

````

# 🔹 NumPy vs Python Lists

NumPy arrays are:
- Faster ✅
- Use less memory ✅
- Support vectorized operations ✅

---

````python
import time

# Python list
lst = list(range(1000000))
start = time.time()
[ x * 2 for x in lst ]
print("List time:", time.time() - start)

# NumPy array
arr = np.arange(1000000)
start = time.time()
arr * 2
print("NumPy time:", time.time() - start)
````
