## 🏗  3: Core Data Structures

### 3.1 `pd.Series`
A one-dimensional labeled array capable of holding any data type.

````python
# Create a Series from a list of values
import pandas as pd
data = [10, 20, 30]
s = pd.Series(data)
print(s)
````

## 🔍5: Exploring a DataFrame

Use the following **built-in methods** to quickly inspect and understand the structure and summary of your `DataFrame`:

---

### 📄 Preview Data
| Method         | Description                            |
|----------------|----------------------------------------|
| `df.head(n)`   | Returns first `n` rows (default = 5)   |
| `df.tail(n)`   | Returns last `n` rows (default = 5)    |

---

### 📊 Summary Information
| Method            | Description                                           |
|-------------------|-------------------------------------------------------|
| `df.info()`       | Overview: index range, column names, non-null counts |
| `df.describe()`   | Statistical summary of numerical columns              |

---

### 🧱 Structure & Metadata
| Attribute        | Description                      |
|------------------|----------------------------------|
| `df.columns`     | List of column names              |
| `df.index`       | Index range of the DataFrame      |
| `df.dtypes`      | Data types of each column         |

---

✅ **Best Practice:** Use these methods as your **first step** when loading a new dataset. They help detect missing values, data types, column structures, and general quality of the data.



<br>

````python
# Shows summary of DataFrame: columns, data types, and non-null counts
df.info()
````

````python
# Shows basic statistics for each numeric column
df.describe()
````

````python
# Shows how many missing values are in each column
df.isnull().sum()
````

## 🎯 6: Practice Use Cases
Here are a few real-world analytics examples you can perform using this dataset.

**🔹 Use Case 1: Total sales per category**

````python
df.groupby('Category')['Total'].sum()
````

**🔹 Use Case 2: Most purchased product**

````python
# Shows total quantity sold for each product, sorted from highest to lowest
df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
````

**🔹 Use Case 3: Filter orders with Total > 1000**

````python
print(df.columns)
````

````python
# Filters rows where the 'Total' value is greater than 1000
df[df['Total'] > 1000]
````

**🔹 Use Case 4: Orders from New York**

````python
# Filters rows where the CustomerCity is 'New York'
df[df['CustomerCity'] == 'New York']
````

## ✂️ 11: Data Selection and Filtering

- Access column: `df['A']`
- Access row by label: `df.loc[0]`
- Access row by position: `df.iloc[0]`
- Boolean filtering: `df[df['A'] > 2]`

````python
# Examples
print(df['A'])  # Series
````

````python
print(df.loc[1])  # row with index 1
````

````python
print(df.iloc[2])  # third row
````

````python
print(df[df['A'] > 3])  # filtering
````

## 🧹 12: Basic Data Cleaning

- Drop missing: `df.dropna()`
- Fill missing: `df.fillna(0)`
- Rename: `df.rename(columns={'A':'Alpha'})`
- Change dtype: `df['A'] = df['A'].astype(float)`

````python
# Cleaning Example
df_clean = pd.DataFrame({'X':[1,None,3], 'Y':[4,5,None]})
print(df_clean.dropna())
print(df_clean.fillna(0))
````

## 🔄 13: DataFrame Operations

- Arithmetic: `df + 1`
- Sorting: `df.sort_values('A', ascending=False)`
- Apply function: `df['A'].map(lambda x: x*2)

```
# This is formatted as code
```

`

````python

# Create a simple DataFrame
data = {
    'A': [5, 2, 8],
    'B': [10, 20, 30]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

````

````python

# ➕ Arithmetic: Add 1 to every value in the DataFrame
print("\nAdd 1 to all values:")
print(df + 1)

````

````python
# 🔽 Sorting: Sort rows by column 'A' in descending order
print("\nSort by column 'A' (descending):")
print(df.sort_values('A', ascending=False))

````

````python
# 🎯 Apply function: Double the values in column 'A' using map()
print("\nDouble the values in column 'A':")
df['A_doubled'] = df['A'].map(lambda x: x * 2)
print(df)
````
