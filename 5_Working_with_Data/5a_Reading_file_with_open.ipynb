## 📄 What is File Handling?

Imagine you’re opening a notebook (the physical one!) to read your class notes. In programming, Python also needs to “open” files to read or write information.

**Why read files?**
- Load saved data
- Read configuration files
- Open logs, documents, or CSVs

We'll use Python’s `open()` function to achieve this.

> 📝 **Note**: You **don’t need to understand the logic** of the code below right now.  
> It simply downloads a sample file from the internet so the upcoming examples work properly.  
> You'll learn how to write such code later in your Python journey. For now, just run it and move on. ✅


````python
import requests
import pandas as pd

filename_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%204/data/example1.txt"
local_filename = "example1.txt"

try:
    response = requests.get(filename_url)
    response.raise_for_status()  # Raise an exception for bad status codes

    with open(local_filename, "wb") as f:
        f.write(response.content)

    print(f"Downloaded {filename_url} to {local_filename}")

except requests.exceptions.RequestException as e:
    print(f"Error downloading the file: {e}")
````

---

````python
# 🔓 Opening a file in read mode
file = open("example1.txt", "r")  # "r" means read mode

# 📖 Reading the entire content
content = file.read()
print(content)

# ❌ Always remember to close the file
file.close()
````

### 🧠 Explanation:
- `"example1.txt"` is the file name (must be in same folder).
- `"r"` stands for "read" mode.
- `file.read()` reads the whole file.
- `file.close()` is important! It frees system resources.

⚠️ **Warning**: If the file doesn't exist, Python will raise an error.

````python
# 🚫 Reading again will return nothing — file pointer is at the end!
file = open("example1.txt", "r")
print(file.read())
print(file.read())  # ⚠️ This will print nothing!
file.close()
````

````python
# 🔁 Read file line by line
file = open("example1.txt", "r")
for line in file:
    print(line.strip())  # Removes trailing newline
file.close()
````

### 🔍 Why line-by-line?
- More memory-efficient for big files
- Easy to process each line separately

````python
# ✅ Automatically closes file even if an error occurs
with open("example1.txt", "r") as file:
    content = file.read()
    print(content)
````

### 🧼 Clean Code Tip
Using `with open()` is safer and preferred:
- Handles closing the file automatically.
- Cleaner syntax for small scripts and large programs.

## 📊 Mini Project: Build a Student Report from File

Suppose you have a file `marks.txt` with contents like:

```
Ali, 85
Sara, 92
Zain, 78
```

**Your task:**
- Read the file
- Print each student’s name and score
- Calculate and print the class average


⚠️ **Note**: The file used in this example doesn’t actually exist—it’s just for practice. If you try to run the code as-is, Python will show an error. Check the solution below and compare it with your own code.

````python
# write Your code here

````

<details>
<summary><strong>✅ Solution</strong></summary>

```python
total = 0
count = 0
with open("marks.txt", "r") as file:
    for line in file:
        name, score = line.strip().split(',')
        score = int(score)
        print(f"{name.strip()} scored {score}")
        total += score
        count += 1
print("Class average:", total / count)
```
</details>
