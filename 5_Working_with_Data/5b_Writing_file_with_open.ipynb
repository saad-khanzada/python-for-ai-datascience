## 🗒️ What Does Writing to a File Mean?

Imagine saving your notes in a diary. Similarly, Python can save text to a file. You can choose to *overwrite*, *append*, or *read+write* using different file modes.

## 🛠️ File Modes Overview

| Mode | Description                     |
|------|---------------------------------|
| `w`  | Write (overwrites existing)     |
| `a`  | Append (adds to end of file)    |
| `r+` | Read and Write (no truncate)    |
| `w+` | Write and Read (overwrites)     |

````python
# 📝 Write to a file (overwrites if exists)
file = open("notes.txt", "w")
file.write("Hello, this file was created using Python!\n")
file.write("Each write adds text.")
file.close()
````

⚠️ **Warning**: Opening a file in `w` mode will erase previous contents without warning.

````python
# ➕ Append to an existing file
file = open("notes.txt", "a")
file.write("\nThis line is added later.")
file.close()
````

````python
# 👓 Let's read the content to verify
with open("notes.txt", "r") as file:
    print(file.read())
````

````python
# 🔄 Read and write without overwriting
file = open("notes.txt", "r+")
print("Before write:")
print(file.read())
file.seek(0)
file.write("Updated Start\n")
file.close()
````

````python
# 🚨 w+ overwrites + allows reading
file = open("rewrite.txt", "w+")
file.write("Starting fresh\n")
file.seek(0)
print(file.read())
file.close()
````

## 📋 Recap Table

| Mode | Action                        | Overwrites? | Allows Read? |
|------|-------------------------------|-------------|----------------|
| `w`  | Write only                    | ✅ Yes       | ❌ No           |
| `a`  | Append only                   | ❌ No        | ❌ No           |
| `r+` | Read and Write                | ❌ No        | ✅ Yes          |
| `w+` | Write and Read (new content)  | ✅ Yes       | ✅ Yes          |

For more file modes (like `rb`, `wb`, `ab` for binary files), check the [Python I/O documentation](https://docs.python.org/3/library/functions.html#open).
