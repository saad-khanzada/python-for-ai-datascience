## 1️⃣ Introduction to Sets

##### What is a Set  <a id='intro'></a>

A **set** in Python is an **unordered** collection of **unique** elements. Think of it like a bag of items where duplicates are ignored and order does not matter.

**Real-life analogy:**
- A set is like a list of students who attended a workshop — if the same student signed in twice, you still only count them once.

Why use sets?
- Remove duplicates quickly.
- Fast membership checks (`in`).
- Useful for mathematical-style operations: union, intersection, difference.


### Creating a Set

```python
# Creating a set of unique fruits
fruits = {"apple", "banana", "cherry", "apple"}  # Duplicate "apple" is removed
print(fruits)
```

Note: You can create a set using the `set()` constructor or curly braces `{}`.

```python
fruits_2 = set(["apple", "banana", "cherry", "apple"])

print(fruits_2)
```

#### 📌 Key Points:

- ✅ **Sets do not allow duplicate values.**  
- 🔄 **Unordered** → No **indexing or position-based access**.

### 2️⃣ Accessing Set Elements

Sets do **not** support **indexing** , but we can loop through elements.

```python
fruits = {"apple", "banana", "cherry", "apple"}

# Looping through a set
for fruit in fruits:
    print(fruit)
```

### 3️⃣ Modifying a Set

We can `len`, `add`, `remove`, `in`,  `discard`, `pop`, `clear`   elements.

```python
fruits = {"apple", "banana", "cherry", "apple"}
print("Original set: ", fruits)

# Getting the length of a set
print("length of a set:", len(fruits))  # Output: 3 as apple are duplicate value


# Adding elements
fruits.add("orange")
print("Adding orange: ", fruits)

# Removing elements
fruits.remove("banana")  # Removes "banana", raises an error if not found
print("Removing banana: ", fruits)

# Checking membership
print("Checking membership: ",  "banana" in fruits)  # Output: False as banana is removed in above method

# Discarding an element (no error if element is missing)
fruits.discard("grape")  # No error
print("Discarding grape: ", fruits)


# Removing a random element
fruits.pop()
print("Removing a random element: ", fruits)

# Clearing all elements
fruits.clear()
print("Clearing all elements:  ", fruits)





```

**Simple notes:**
- Use `discard` when you're not sure the element exists (it won't error). Use `remove` if you want an error when element missing (helps catch bugs).

### 4️⃣ Set Operations

Sets support **mathematical operations** like **union, intersection** , and **difference**.

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Union: Combines all elements (removes duplicates)
print(A | B)  # Output: {1, 2, 3, 4, 5, 6}

# Intersection: Common elements
print(A & B)  # Output: {3, 4}

# Difference: Elements in A but not in B
print(A - B)  # Output: {1, 2}

# Symmetric Difference: Elements in either A or B but not both
print(A ^ B)  # Output: {1, 2, 5, 6}
```

```python
# Student Course Enrollment using sets
English = {'Ali', 'Saad', 'Hassan', 'Maya'}
Urdu = {'Saad', 'Omar', 'Maya', 'Zain'}

print("English:", English)
print("Urdu ", Urdu )

both = English & Urdu
only_a = English - Urdu
only_b = Urdu- English
either = English | Urdu

print("\nStudents in both courses:", both)
print("Students only in  English Course:", only_a)
print("Students only in  Urdu Course:", only_b)
print("All students (either course English or Urdu  ):", either)
```

### 6️⃣ Frozen Sets (Immutable Sets)

A **frozen set** is a **read-only version** of a set. it can't be modify.

```python
# Creating a frozen set
frozen_A = frozenset(A)
print(frozen_A)

```

```python
# Trying to modify it (will raise an error)
frozen_A.add(5)  # ❌ Error
```
