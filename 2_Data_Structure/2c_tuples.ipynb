## What is a Tuple?


A **tuple** in Python is an **ordered, immutable collection** of items.
- Ordered → items have a fixed position (like a list).  
- Immutable → cannot change after creation.  


## Creating Tuples

````python
# Creating a tuple of fruits
fruits = ("Apple", "Banana", "Cherry")
print(fruits)
````

````python
# Tuple with mixed data types
mixed = (1, "Hello", 3.14, True)
print(mixed)
````

````python
# Single-element tuple requires a comma!
single = (42,)
print(single, type(single))
````

#### 📌 Key Points:
### Tuples in Python

- **Tuples are defined using parentheses** `()`
- **They can store multiple data types** (e.g., strings, numbers)
- **Tuples are immutable** (cannot be changed after creation)


___

## Accessing Tuples

````python
fruits = ("Apple", "Banana", "Cherry")

# Indexing
print(fruits[0])   # First element
print(fruits[-1])  # Last element

# Slicing
print(fruits[0:2]) # From index 0 to 1
````

## Tuple Operations

````python
# Concatenation
tuple1 = (1, 2, 3)
tuple2 = (4, 5)
print(tuple1 + tuple2)

# Repetition
print(("Hi!",) * 3)

# Membership
print("Banana" in fruits)
````

## Tuple Methods

````python
numbers = (1, 2, 2, 3, 4, 2)

print(numbers.count(2))  # Count occurrences
print(numbers.index(3))  # Index of element
````

## Packing and Unpacking

**Packing** = grouping values together .

**Unpacking** = extracting those grouped values back into separate variables.

````python
# Packing
coordinates = (10, 20, 30)

# Unpacking
x, y, z = coordinates
print(f"x={x}, y={y}, z={z}")
````

## Tuples vs Lists


| Feature           | Tuple                | List                 |
|-------------------|----------------------|----------------------|
| Mutability        | Immutable            | Mutable              |
| Syntax            | `(1, 2, 3)`          | `[1, 2, 3]`          |
| Performance       | Faster (fixed size)  | Slower (resizable)   |
| Use cases         | Fixed data, keys     | Dynamic collections  |

## Summary


- Tuples are **immutable, ordered collections**.  
- Use `()` for tuples, `[ ]` for lists.  
- Support indexing, slicing, concatenation, repetition.  
- Useful methods: `.count()`, `.index()`.  
- Tuples can be **packed/unpacked** easily.  
- Ideal for fixed data (e.g., coordinates, keys in dictionaries).
