# 1️: Introduction to Lists

A list in Python is an ordered collection of items that can store multiple values in a single variable. Lists are mutable, meaning their elements can be changed after creation.

<hr>

<br> <br>

```python
# Creating a simple list of fruits
fruits = ["Apple", "Banana", "Cherry"]
print(fruits)  # Output: ['Apple', 'Banana', 'Cherry']
```

<hr>

### Accessing Elements in a List

You can access elements in a list using indexing (starting from 0).

```python
# Accessing elements
print(fruits[0])  # Output: Apple
print(fruits[1])  # Output: Banana
print(fruits[-1]) # Output: Cherry (negative indexing)
```

### Modifying Lists

Lists are mutable, meaning you can change their elements.

```python
# Modifying an element
fruits[1] = "Mango"
print(fruits)  # Output: ['Apple', 'Mango', 'Cherry']
```

4️⃣ Adding and Removing Elements \
<code> append()</code> Adds an item at the end \
<code>insert()</code> Adds an item at a specific position\
<code>pop()</code> Removes the last item or an item at a given index\
<code>remove()</code> Removes a specific value

<br><br>

```python
# Adding elements
fruits.append("Orange")   # Adds 'Orange' at the end

print(fruits)

```

<hr>

```python
fruits.insert(1, "Grapes") # Inserts 'Grapes' at index 1

print(fruits)
```

<hr>

```python
# Removing elements
fruits.pop()   # Removes last element

print(fruits)
```

<hr>

```python
fruits.remove("Apple")  # Removes 'Apple'

print(fruits)
```

 ## Slicing Lists

You can extract a portion of a list using slicing.

```python
numbers = [10, 20, 30, 40, 50]
print(numbers[1:4])  # Output: [20, 30, 40]
print(numbers[:3])   # Output: [10, 20, 30]
print(numbers[2:])   # Output: [30, 40, 50]
```
