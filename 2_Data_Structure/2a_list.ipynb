## 1️⃣ Introduction  
A **list** in Python is an ordered, mutable collection of elements.  
Lists can store numbers, strings, or even other lists.

## 3️⃣ Indexing and Slicing

````python

fruits = ["apple", "banana", "cherry", "date"]

# Indexing
print(fruits[0])   # First element
print(fruits[-1])  # Last element

# Slicing
print(fruits[1:3]) # Elements from index 1 to 2
print(fruits[:2])  # First two elements
print(fruits[2:])  # From index 2 to end
````

___

## 4️⃣ Modifying Lists

<code> append()</code> Adds an item at the end \
<code>insert()</code> Adds an item at a specific position\
<code>remove()</code> Removes a specific value\
<code>pop()</code> Removes the last item or an item at a given index

````python

fruits = ["apple", "banana", "cherry"]
fruits[1] = "blueberry"   # Change element at index 1
print(fruits)

fruits.append("date")     # Add element to end
print(fruits)

fruits.insert(1, "kiwi")  # Insert at index 1
print(fruits)

fruits.remove("apple")    # Remove by value
print(fruits)

popped_item = fruits.pop()  # Remove last element
print(fruits)
print("Popped:", popped_item)
````

## 5️⃣ Common List Methods

````python

numbers = [5, 2, 9, 1, 7]

numbers.sort()   # Sort in ascending order
print("Sorted:", numbers)

numbers.reverse()   # Reverse order
print("Reversed:", numbers)

print("Count of 2:", numbers.count(2))  # Count occurrences
print("Index of 7:", numbers.index(7))  # Find index
````

___

## 6️⃣ Iterating Over Lists

````python

fruits = ["apple", "banana", "cherry"]

# For loop
for fruit in fruits:
    print(fruit)

# Using enumerate for index + value
for index, fruit in enumerate(fruits):
    print(index, fruit)
````

___

## 7️⃣ Nested Lists

````python

matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

print(matrix[0])      # First row
print(matrix[1][2])   # Element at row 2, column 3
````
