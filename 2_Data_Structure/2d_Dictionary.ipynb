### 1️⃣ Introduction to Dictionaries

#### What is a Dictionary?

A dictionary in Python is a collection of key-value pairs. It allows you to store and retrieve data efficiently.

##### Creating a Dictionary

```python
# Creating a dictionary of student information
student = {
    "name": "saad",
    "age": 20,
    "grade": "A",
    "city": "Islamabad"
}
print(student)
```

#### 📌 Key Points:

- **Keys** must be **unique** and **immutable** (e.g., strings, numbers, tuples).  
- **Values** can be of **any data type**.  
- **Dictionaries are unordered** in Python versions **before 3.7** and **ordered in Python 3.7+**.

### 2️⃣ Accessing Dictionary Values

We can access dictionary values using their  <code> keys <code\>

```python
# Accessing values
print(student["name"])  # Output: saad
print(student["grade"]) # Output: A
```



 Alternatively, use  `.get()`  to **avoid errors** if the key does not exist.


```python
# Using .get() to access values
print(student.get("age"))        # Output: 20
print(student.get("email", "Not found"))  # Output: Not found
```

### 3️⃣ Modifying a Dictionary

Dictionaries are mutable, meaning we can change their values.

```python
# Updating values
student["age"] = 21
print(student)  # Updated age to 21
```

We can also *add new key-value pairs*

```python
# Adding a new key-value pair
student["email"] = "saad@example.com"
print(student)
```

### 4️⃣ Removing Elements

We can remove elements using different methods:

```python
# Removing a specific key using pop()
student.pop("grade")
print(student)  # "grade" key is removed

```

<hr>

```python
# Using del keyword
del student["city"]
print(student)  # "city" key is removed

```

<hr>

```python
# Using clear() to remove all elements
student.clear()
print(student)  # Output: {}
```

### 5️⃣ Looping Through a Dictionary

We can iterate through a dictionary using a `for` loop.

```python
student = {"name": "saad", "age": 20, "grade": "A"}

# Loop through keys
for key in student:
    print(key)

```

<hr>

```python
# Loop through values
for value in student.values():
    print(value)

```

<hr>

```python
# Loop through key-value pairs
for key, value in student.items():
    print(f"{key}: {value}")

```

### 6️⃣ Dictionary Methods

Dictionaries have several useful methods:

```python
# Checking if a key exists
print("name" in student)  # Output: True


```

<hr>

```python
# Getting all keys
print(student.keys())  # Output: dict_keys(['name', 'age', 'grade'])

# Getting all values
print(student.values())  # Output: dict_values(['saad', 20, 'A'])

```

<hr>

```python
# Getting all key-value pairs
print(student.items())  # Output: dict_items([('name', 'saad'), ('age', 20), ('grade', 'A')])
```

### 7️⃣ Nested Dictionaries

Dictionaries can store *other dictionaries* inside them.

```python
students = {
    "saad": {"age": 20, "grade": "A"},
    "Sara": {"age": 22, "grade": "B"}
}
print(students["saad"]["grade"])  # Output: A
```
