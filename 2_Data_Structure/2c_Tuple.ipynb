#  Introduction to Tuples


## Tuple

A tuple in Python is an ordered, immutable collection  of items. Unlike lists, tuples cannot be modified after creation. Tuples are useful when you need to store data that should remain unchanged.

#### Creating a Tuple

```python
# Creating a tuple of fruits
fruits = ("Apple", "Banana", "Cherry")
print(fruits)  # Output: ('Apple', 'Banana', 'Cherry')
```

#### 📌 Key Points:
### Tuples in Python

- **Tuples are defined using parentheses** `()`
- **They can store multiple data types** (e.g., strings, numbers)
- **Tuples are immutable** (cannot be changed after creation)

 ## Accessing Elements in a Tuple

We can access tuple elements using indexing (starting from 0).




```python
# Accessing elements
print(fruits[0])  # Output: Apple
print(fruits[1])  # Output: Banana
print(fruits[-1]) # Output: Cherry (negative indexing)
```


### Why Use Tuples?

- **Prevent accidental modifications** (tuples are immutable)
- **Faster than lists** (due to their fixed size)
- **Can be used as dictionary keys** (because hash value remains constant)

## Tuple Operations

Even though tuples are immutable, we can perform some operations:



#### Tuple Concatenation

```python
# Joining two tuples
fruits = ("Apple", "Banana", "Cherry")
vegetables = ("Carrot", "Potato", "Tomato")
food = fruits + vegetables
print(food)  # Output: ('Apple', 'Banana', 'Cherry', 'Carrot', 'Potato', 'Tomato')
```

<hr>

#### Tuple Repetition

```python
# Repeating elements
repeat_tuple = ("Hello",) * 3
print(repeat_tuple)  # Output: ('Hello', 'Hello', 'Hello')
```

##  Slicing Tuples

We can extract specific portions of a tuple using slicing.

```python
numbers = (10, 20, 30, 40, 50)
print(numbers[1:4])  # Output: (20, 30, 40)
print(numbers[:3])   # Output: (10, 20, 30)
print(numbers[2:])   # Output: (30, 40, 50)
```

## Looping Through Tuples

Tuples can be iterated using a for loop.

```python
# Using a loop to print elements
for fruit in fruits:
    print(fruit)
```

##  Tuple Methods

Tuples have a few useful built-in methods.

```python
# Count occurrences of an element
numbers = (1, 2, 3, 2, 4, 2)
print(numbers.count(2))  # Output: 3

# Find the index of an element
print(numbers.index(3))  # Output: 2
```
