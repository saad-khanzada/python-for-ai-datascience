#### 🔵 Float Type (`float`)

A **float** (short for "floating-point number") is a number that **includes a decimal point**.

Examples: `3.14`, `-0.5`, `0.0`

##### 🔹 Common Uses:
- Representing measurements, percentages, or averages
- Financial calculations (e.g., prices, tax rates)


````python
# Floats include decimal numbers
float_number = 10.5
negative_float = -3.14
zero_float = 0.0

# Printing float values
print("Positive Float:", float_number)
print("Negative Float:", negative_float)
print("Zero Float:", zero_float)
````

#### 🔄 Typecasting: Converting Integers to Floats

**Typecasting** means converting one data type into another.

To convert an integer to a float in Python, use the built-in `float()` function.

##### 🧠 Why do it?
- When you need **decimal precision**
- For **mathematical operations** involving both integers and floats

##### 🛠️ Syntax:
```python
float(10)  # Result: 10.0


````python
# Converting an integer to a float
int_value = 7
int_to_float = float(int_value)

# Printing the conversion result
print("Original Integer:", int_value)
print("Converted to Float:", int_to_float)
````

#### 🔄 Typecasting: Converting Floats to Integers

You can convert a **float** (decimal number) into an **integer** using the `int()` function.

⚠️ This **removes the decimal part** — it does **not round**, it **truncates**.

##### 🛠️ Syntax:
```python
int(4.9)  # Result: 4


````python
# Converting a float to an integer
float_value = 12.99
float_to_int = int(float_value)  # Removes the decimal part (truncation)

# Printing the conversion result
print("Original Float:", float_value)
print("Converted to Integer:", float_to_int)
````

#### 🔤 Typecasting: Converting Integers and Floats to Strings

You can convert both **integers** and **floats** into **strings** using the `str()` function.

This is useful when you want to:
- Combine numbers with text
- Display numbers in a user-friendly way

##### 🛠️ Syntax:
```python
str(25)     # "25"
str(3.14)   # "3.14"


````python
# Converting an integer to a string
int_value = 25
int_to_str = str(int_value)

# Converting a float to a string
float_value = 7.89
float_to_str = str(float_value)

# Printing the results
print("Integer as String:", int_to_str)
print("Float as String:", float_to_str)
````

#### 🔁 Typecasting: Converting Numbers to Booleans

You can convert any number to a Boolean using the `bool()` function.

##### 🔹 Rules:
- `0` or `0.0` → `False`
- Any **non-zero** value (positive or negative) → `True`

##### 🛠️ Syntax:
```python
bool(0)      # False
bool(7)      # True
bool(-3.5)   # True


````python
# Any non-zero number is considered True, and 0 is considered False in Boolean conversion
int_value = 10
zero_value = 0
float_value = -5.6

# Converting to Boolean
bool_from_int = bool(int_value)
bool_from_zero = bool(zero_value)
bool_from_float = bool(float_value)

# Printing the results
print("Boolean from Integer (10):", bool_from_int)
print("Boolean from Zero (0):", bool_from_zero)
print("Boolean from Float (-5.6):", bool_from_float)
````
