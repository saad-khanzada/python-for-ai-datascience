## 🧮 What is a Variable in Python?

Before we work with expressions, we first need to understand **variables**.


A **variable** is a name used to **store data** in memory so it can be reused later.  
In Python, you create a variable by using the `=` sign to assign a value.

📌 **Example:**

```python
x = 10
name = "Ali"


````python
# Variables store values for reuse in calculations
x = 20
y = 5
sum_result = x + y
product_result = x * y

# Printing variable values and results
print("Value of x:", x)
print("Value of y:", y)
print("Sum of x and y:", sum_result)
print("Product of x and y:", product_result)
````

# Expressions in Python

## 🧠 What is an Expression?

In Python, an **expression** is any combination of **values, variables, and operators** that Python can evaluate (compute) into a result.

For example:

```python
2 + 3
a * 5
x / y - 1


````python
# Expressions are combinations of values and operations that produce a single result
result = 5 + 3 * 2  # Multiplication (*) happens first, then addition (+)
print("Expression Result (5 + 3 * 2):", result)
````

## Integer Division (//)

````python
# Integer division (//) discards the fractional part
num1 = 15
num2 = 4
int_div_result = num1 // num2  # 15 / 4 = 3.75, but // removes .75

print("Integer Division Result (15 // 4):", int_div_result)
````

### ✅ Operator Precedence (BODMAS Rule)


## 🔢 Operator Precedence in Python

Python follows the **BODMAS rule** when evaluating expressions:

- **B**rackets `()`
- **O**rders: Exponents `**`
- **D/M**: Division `/`, Floor Division `//`, Modulus `%`
- **A/S**: Addition `+`, Subtraction `-`

### 🧪 Example:



````python
result = 3 + 4 * 2  #answer will be 11 (Not 14)
print(result)
````

Multiplication happens before addition.

Use () to control the order:

````python
result = (3 + 4) * 2  # Output: 14
result
````

````python
# Python follows BODMAS (Brackets, Orders, Division/Multiplication, Addition/Subtraction)
expr1 = (2 + 3) * 4  # Brackets first, then multiplication
print("Expression 1 ((2 + 3) * 4):", expr1)
````

<br><br>

````python
expr2 = 10 - 3 * 2  # Multiplication first, then subtraction
print("Expression 2 (10 - 3 * 2):", expr2)

````

<br><br>

````python
expr3 = 100 / 5 + 2  # Division first, then addition
print("Expression 3 (100 / 5 + 2):", expr3)
````
