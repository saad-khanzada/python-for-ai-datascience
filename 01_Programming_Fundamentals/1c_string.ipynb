##Introduction

Strings in Python are an ordered sequence of characters used to store and manipulate text data. Since strings are sequences, we can perform various operations on them, such as slicing, indexing, concatenation, and replication.


## 🎯 Learning Objectives  
By the end of this notebook, you will be able to:  
- Create and manipulate strings in Python.  
- Access string characters using indexing and slicing.  
- Apply common string methods (e.g., `strip`, `split`, `replace`).  
- Format strings using f-strings and `.format()`.  
- Use escape characters for special formatting.  


## 1️⃣ Creating Strings

```python

# Strings can be created using single or double quotes
string1 = 'Hello'
string2 = "World"
print(string1, string2)
```

## 2️⃣ Indexing and Slicing


Indexing allows us to access individual characters in a string using positive (left-to-right) or negative (right-to-left) indices.

```python

text = "Python"
# Indexing
print(text[0])   # First character
print(text[-1])  # Last character

# Slicing
print(text[0:3])   # First three letters
print(text[2:])    # From index 2 to end
```

## 3️⃣ Common String Operations

```python

# Concatenation
greeting = "Hello" + " " + "Python"
print(greeting)

# Repetition
laugh = "Ha" * 3
print(laugh)

# Length of string
print(len(greeting))
```

## 4️⃣ Useful String Methods

```python

sample = "   Python is Fun!   "

print(sample.strip())      # Remove spaces
print(sample.lower())      # Lowercase
print(sample.upper())      # Uppercase
print(sample.replace("Fun", "Powerful"))

words = sample.split()     # Split into list
print(words)

joined = "-".join(words)   # Join with '-'
print(joined)
```

**Finding Substrings in a String**

The find() method returns the index of the first occurrence of a substring or -1 if not found.

```python
text = "Learning Python is exciting!"

index = text.find("Python")
print("Index of 'Python':", index)

# Searching for a non-existing word
not_found = text.find("Java")
print("Index of 'Java':", not_found)  # Returns -1
```

## 5️⃣ String Formatting

```python

name = "Saad"
age = 21

# f-string
print(f"My name is {name} and I am {age} years old.")

# format method
print("My name is {} and I am {} years old.".format(name, age))
```

## 6️⃣ Escape Characters


Escape sequences () help format strings by adding special characters like new lines <code> (\n) <code\> and <code> tabs (\t).<code\>

```python

text = "Hello\nWorld"   # \n = newline
print(text)

text2 = "Tab\tSpace"
print(text2)
```

## 7️⃣ Practical Examples

```python

# Example: Extract username from email
email = "student@example.com"
username = email.split("@")[0]
print("Username:", username)

# Example: Clean user input
raw_text = "   Python Programming   "
clean_text = raw_text.strip().lower()
print("Cleaned text:", clean_text)
```


## 📌 Summary  
- Strings are sequences of characters defined with quotes.  
- Use **indexing and slicing** to access parts of strings.  
- Common operations: concatenation, repetition, length.  
- Useful methods: `strip()`, `split()`, `join()`, `replace()`, `upper()`, `lower()`.  
- Formatting with **f-strings** or `.format()`.  
- Escape characters like `\n` and `\t` add special formatting.
