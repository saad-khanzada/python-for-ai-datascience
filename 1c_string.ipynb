
  ### Understanding Strings in Python

Strings in Python are an ordered sequence of characters used to store and manipulate text data. Since strings are sequences, we can perform various operations on them, such as slicing, indexing, concatenation, and replication.

#### Assigning Strings to Variables

 A string is a sequence of characters stored in a variable using single or double quotes.

```python
greeting = "Hello, Python!"
name = 'Alice'

print("Greeting:", greeting)
print("Name:", name)
```

### String Indexing (Accessing Characters)


Indexing allows us to access individual characters in a string using positive (left-to-right) or negative (right-to-left) indices.

```python
word = "Python"

print("First character (Positive index 0):", word[0])  # P
print("Last character (Negative index -1):", word[-1])  # n
print("Third character (Positive index 2):", word[2])  # t
print("Second-last character (Negative index -2):", word[-2])  # o
```

####  String Slicing


Slicing allows us to extract parts of a string using the syntax <code> [start:end:step]<code\>

```python
sentence = "Programming"

print("First 5 characters:", sentence[:5])  # Progr
print("Characters from index 2 to 8:", sentence[2:8])  # ogram
print("Every second character:", sentence[::2])  # Pormig
print("Reverse string:", sentence[::-1])  # gnimmargorP
```

#### . Changing String Case

Methods like upper() and lower() change the case of characters in a string.

```python
message = "Python is Fun!"

print("Uppercase:", message.upper())  # Converts to uppercase
print("Lowercase:", message.lower())  # Converts to lowercase
print("Title Case:", message.title())  # Capitalizes each word
```

####  Finding Substrings in a String


The find() method returns the index of the first occurrence of a substring or -1 if not found.

```python
text = "Learning Python is exciting!"

index = text.find("Python")
print("Index of 'Python':", index)

# Searching for a non-existing word
not_found = text.find("Java")
print("Index of 'Java':", not_found)  # Returns -1
```

#### Replacing Parts of a String

The replace() method replaces one substring with another in a string.

```python
sentence = "I love Java programming."

new_sentence = sentence.replace("Java", "Python")  # Replacing 'Java' with 'Python'
print("Original Sentence:", sentence)
print("Modified Sentence:", new_sentence)
```

#### Splitting and Joining Strings

 The split() method divides a string into a list, while join() combines a list into a string.

```python
quote = "Python is simple yet powerful"

# Splitting a string into words
words = quote.split()  # Default splits by spaces
print("Words List:", words)

# Joining words back into a string
joined_text = "-".join(words)
print("Joined String:", joined_text)  # Python-is-simple-yet-powerful
```

####  Stripping Whitespace from a String


The strip() method removes leading and trailing spaces from a string.

```python
raw_text = "   Hello, World!   "
clean_text = raw_text.strip()  # Removes spaces

print("Before Strip:", repr(raw_text))  # Using repr() to show spaces
print("After Strip:", repr(clean_text))
```
